```python
def delimiter(word):
    delimiter_list = [' ', '\n', '[', ']', '"', '.', ',', 'ECP']
    if word in delimiter_list:
        return False
    return True
```

```python
import jieba.posseg as pseg

text_file = open('customer_data.txt', 'r', encoding='utf-16le')
output_file = open('output.txt', 'w', encoding='utf-8')
for line in text_file.readlines():
    line_split = line.split('\t')
    words = pseg.cut(line_split[3])
    output_line = ''
    for word, flag in words:
        if(delimiter(word)):
#             print('%s %s' % (word, flag))
            output_line += word + ' '
    print(output_line.replace(" ", ""))
    output_file.write(output_line.replace(" ", ""))
text_file.close()
output_file.close()
```

```python
text_file = open('customer_data.txt', 'r', encoding='utf-16le')
jieba_file = open('jieba.txt', 'w', encoding='utf-8')
corpus = []
for line in text_file.readlines():
    line_split = line.split('\t')
    words = pseg.cut(line_split[3])
    sentence = ''
    for word, flag in words:
        if(delimiter(word)):
#             print('%s %s' % (word, flag))
#             print(word + flag)
#             jieba_file.write(word + flag + '\n')
            sentence += (word + ' ')
#     print(sentence)
    corpus.append(sentence)
print(corpus)
text_file.close()
jieba_file.close()
```

```python
import pandas as pd
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
 
vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(corpus))
terms = vectorizer.get_feature_names()

sums = tfidf.sum(axis=0)

# connecting term to its sums frequency
data = []
for col, term in enumerate(terms):
    data.append( (term, sums[0,col] ))

ranking = pd.DataFrame(data, columns=['term','score'])
print(ranking.sort_values('score', ascending=False))

```

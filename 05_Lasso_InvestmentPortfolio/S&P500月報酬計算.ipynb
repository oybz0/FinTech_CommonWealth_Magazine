# 月報酬：[((Pt+1)-Pt)/Pt]**12
Pt+1 : 每個月一號的指數   
Pt：下個月一號的指數

```python
import datetime
x=(datetime.date.today())
for i in range(1,121):
    if x.month > 1 :
        x=x.replace(month=x.month-1,day=1)
    else:
        x=x.replace(year=x.year-1,month=12,day=1)
    print(x)
```

```python
import pandas as pd
df1= pd.read_csv('/Users/chenjessie/Desktop/經濟預測模型/s&p500.csv')
df1.head()
```

```python
df1['sp500'] = 0
df1['condition'] = 0
df1['predict'] = 0
for i in range(df1.shape[0]-1):
    rate = float(format(((df1.iloc[i+1]['Close'] - df1.iloc[i]['Close'])/ df1.iloc[i]['Close']),'.10f') )
    df1.loc[i,'sp500'] = rate
    
    if rate > 0.07:
        df1.loc[i,'condition'] = '3'
        if i-1>=0:
            df1.loc[i-1,'predict'] = '3'
    elif rate < (-0.07):
        df1.loc[i,'condition'] = '1'
        if i-1>=0:
            df1.loc[i-1,'predict'] = '1'
    else:
        df1.loc[i,'condition'] = '2'
        if i-1>=0:
            df1.loc[i-1,'predict'] = '2'
```
